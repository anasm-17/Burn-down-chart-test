```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import datetime
from datetime import date, timedelta
%matplotlib inline
```

```python
def get_initial_params():
    with open("Initialize_burn_down.txt") as f:
        content = f.readlines()
    flag = int(float(content[0].split(':')[0]))
    days = float(content[1].split(':')[0])
    tasks = float(content[2].split(':')[0])
    return flag,days,tasks
```

```python
def get_tasks_completed():
    with open("Tasks_completed.txt") as f:
        content = f.readlines()
    completed = float(content[0].split(':')[0])
    return completed
```

```python
#project start date
# 0 - project start
# 1 - project ongoing
# 2 - project finished
project_status_flag,Number_of_Days,Number_of_Tasks = get_initial_params()


y_ideal=np.linspace(0,Number_of_Tasks,Number_of_Days)
y_ideal=y_ideal[::-1]
x_ideal=np.linspace(1,Number_of_Days,Number_of_Days)
```

```python
project_status_flag
```

```python
Tasks_completed_today = get_tasks_completed()
```

```python
Tasks_completed_today
```

```python
if (project_status_flag == 0):
    project_start_date=date.today()
    project_status_flag=1
    y=[]
    y.append(Number_of_Tasks)
```

```python
if (project_start_date==date.today()):
    Day=[]
    Day.append(1.0)
    day_iter=0
else:
    str_Day = str((date.today()-project_start_date).strftime('%d-%m-%y').split('-')[0])
    f_Day = float(str_Day)
    if f_Day not in Day:
        Day.append(f_Day)
        day_iter+=1
        y.append(y[day_iter-1])
```

```python
y[day_iter] = y[day_iter-1]-Tasks_completed_today
```

```python
plt.figure(figsize=(15,10),dpi=200)
plt.plot(x_ideal,y_ideal,marker='x', markerfacecolor='blue', markersize=4, color='skyblue', linewidth=2)
plt.plot(Day,y, marker='x', color='red', markersize=4, linewidth=2)
plt.xticks(x_ideal)
plt.xlim(left=1)
plt.xlabel("Number of Days",fontsize=15)
plt.ylabel("Number of Tasks",fontsize=15)
plt.legend(labels=("Ideal Burn Down","Actual Burn Down"),fontsize=15)
plt.show()
```
